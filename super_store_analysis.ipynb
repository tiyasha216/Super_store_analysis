```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("superstore_messy_dataset.csv")
df.head()
```

```python
# df.duplicated().any()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```

```python
df['Discount'].fillna(df['Discount'].median(), inplace=True)
df['Profit'].fillna(df['Profit'].median(), inplace=True)
```

```python
df['City'].fillna(df['City'].mode()[0], inplace=True)
```

```python
df['Sales'].fillna(df['Sales'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Extract month and year

df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
# Profit Margin
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
df['Sales Category'] = pd.cut(df['Sales'],
                             bins=[0,100,500,1000,5000],
                             labels=['Low','Medium','High','Very High'])
```

```python
print("Mean Sales:", df['Sales'].mean())
print("Median Profit:", df['Profit'].median())
print("Std Dev:", df['Sales'].std())
```

```python
# Which category has the highest total sales?
```

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
# Which category is most profitable?
```

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
# Are there loss-making transactions?

df[df['Profit'] < 0].shape[0]
```

```python
# Monthly sales trend

df.groupby('Month')['Sales'].sum().plot(kind='line', title='Monthly Sales Trend')
```

```python
# Which customer segment contributes most?


df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
```

```python
# Which city generates highest sales?

df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
df.to_csv("cleaned_superstore.csv", index=False)
```

```python
# Profit by Region

df.groupby('Region')['Profit'].sum().sort_values()
```
